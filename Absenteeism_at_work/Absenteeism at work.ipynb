```python
plt.figure(figsize=(10,5))
data[data['Son']!=0]['Absenteeism time in hours'].plot.hist(bins=30)
plt.xlabel("'Son'/ Number of children")
plt.ylabel('Absenteeism time in hours')
plt.show()
```

### In the group of people with children, absence in hours tends to be between 0 and 20, and very small amount was absence more than that

```python
plt.figure(figsize=(10,6))
data[data['Son']==0]['Absenteeism time in hours'].plot.hist(bins=30)
plt.xlabel("'Son'/ Number of children")
plt.ylabel('Absenteeism time in hours')
plt.show()
```

### The number of people without children showed more absenteeism than the people with children.

```python
a = sns.FacetGrid(data=data,col='Son')
a.map(plt.hist,'Absenteeism time in hours')
plt.show()
```

### Here all possible options for 'Son' column are shown in their correlation with absence time. We can see that, if you have 0 or 1 children, it is more likely to be absent at work. Probably, because if you have only one child and a mother presumably, noone can sit with child, apart from you and your spouse. Plus being a first time parent can be overwhelming given a parent is still learning the do and donts of parenting. But, if we look further, more children you have, less time you are absent. It means, that they can basically watch each other, and you dont need to stay home to take care of them and they have adopted the rules and day-to-day dos and donts of parenting
